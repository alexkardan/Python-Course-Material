```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {'Category': ['A', 'B', 'C', 'D', 'E'],
        'Value1': [15, 30, 5, 10, 20],
        'Value2': [10, 25, 8, 15, 30],
        'Value3': [5, 10, 15, 20, 25]}

df = pd.DataFrame(data)
```

```python
sns.barplot(x = "Category", y = "Value1", hue = "Category", data = df)
plt.title("Bar Plot")
plt.xlabel("Category")
plt.ylabel("Value1")
plt.show()
```

```python
sns.scatterplot(x = "Value1", y = "Value2", hue = "Category", data = df)
plt.title("Scatter Plot")
plt.xlabel("Value1")
plt.ylabel("Value2")
plt.legend(title = "Category")
plt.show()
```

```python
sns.lineplot(x = "Category", y = "Value1", data = df, marker = "o")
plt.title("Line Plot")
plt.xlabel("Category")
plt.ylabel("Value1")
plt.show( )
```

```python
sns.heatmap(df[["Value1", "Value2", "Value3"]].corr(), annot = True, cmap = "coolwarm")
plt.title("Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue = "Category", diag_kind = "kde")
plt.suptitle("Pairplot", y = 1.02)
plt.show()
```

```python
sns.countplot(x = "Category", hue = "Category", data = df)
plt.title("Count Plot")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
sns.kdeplot(data=df[["Value1", "Value2", "Value3"]], fill = True)
plt.title("KDE Plot")
plt.show()
```

```python
sns.regplot(x = "Value1", y = "Value2", data = df, marker = "o")
plt.title("Regplot")
plt.xlabel("Value1")
plt.ylabel("Value2")
plt.show()
```
